## `plot_sinc.py` script

```python
import matplotlib.pyplot as plt
import math

plt.matplotlib.style.use("ggplot")

def sinc(x):
  """sinc(x) = sin(x)/x function (x in rad)"""
  return math.sin(x)/x if x != 0 else 1
  
# use list comprehensions
# -5π ≤ x ≤ +5π
X = [0.1*math.pi*i for i in range(-50, 51)]
# y = sinc(x)
Y = [sinc(x) for x in X]

plt.plot(X, Y, marker=".")
plt.xlabel("x")
plt.ylabel("y")

# show the graph (may not be necessary in the notebook)
plt.show()

# save plot to file sinc.png: it should show up in the directory where the notebook is located
plt.savefig("sinc.png")
```

## Step-by-step

Import the `matplotlib` package for plotting and the standard library's `math` module:

```python
import matplotlib.pyplot as plt
import math
```

matplotlib gives us different styles for plots. I like "ggplot" style but you should experiment and check out [all the different styles](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html). 

```python
plt.matplotlib.style.use("ggplot")
```

We want to plot the [sinc](https://mathworld.wolfram.com/SincFunction.html) function 
\begin{equation}
\mathrm{sinc}(x) = \begin{cases}
1&\ \text{for}\ x = 0\\
\frac{\sin x}{x}&\ \text{otherwise}
\end{cases}
\end{equation}
so we _define_ it as a Python function:

```python
def sinc(x):
  """sinc(x) = sin(x)/x function (x in rad)"""
  return math.sin(x)/x if x != 0 else 1
```

We will plot over the range –5π ≤ x ≤ 5π with step size 0.1π. We create a list `X` of $x$-values with a list comprehension:

```python
X = [0.1*math.pi*i for i in range(-50, 51)]
```

Now we evaluate the `sinc` function for all values in `X` to generate a list `Y` of `sinc(x)` values: This is the actual function evaluation 

$$
y = \mathrm{sinc}(x)
$$

(For convenience, we create the `Y` list again with a list comprehension.)

```python
Y = [sinc(x) for x in X]
```

**Plot** the $x$ and $y$ coordinates with the [plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) function. We also add axes labels.

```python
plt.plot(X, Y, marker=".")
plt.xlabel("x")
plt.ylabel("y");
```

Finally, save the figure to a file with [savefig()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html). You can choose different formats (PDF, JPG, ...) by choosing a different file ending. The file should appear in the same folder where the notebook is located.

```python
plt.savefig("sinc.png")
```
